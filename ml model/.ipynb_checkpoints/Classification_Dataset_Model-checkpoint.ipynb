```python
import pandas as pd
```

```python
file = "https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data"

names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df = pd.read_csv(file, names = names, header = 0)

df.head()
```

```python
df['doors'].unique()
```

```python
df['persons'].unique()
```

```python
df["buying"].replace({"vhigh": 4, "high": 3, "med": 2, "low": 1}, inplace = True)
df["maint"].replace({"vhigh": 4, "high": 3, "med": 2, "low": 1}, inplace = True)
df["doors"].replace({"2": 2, "3": 3, "4": 4, "5more": 5}, inplace = True)
df["persons"].replace({"2": 2, "4": 4, "more": 6}, inplace = True)
df["lug_boot"].replace({"small": 1, "med": 2, "big": 3}, inplace = True)
df["safety"].replace({"low": 1, "med": 2, "high": 3}, inplace = True)
df["class"].replace({"unacc": 1, "acc": 2, "good": 3, "vgood": 4}, inplace = True)
```

```python
df.head()
```

```python
array = df.values
X = array[:,:-1]
Y = array[:,-1]
```

```python
models = []

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

log = LogisticRegression()
models.append(('LR', log))

nb = GaussianNB()
models.append(('NB', nb))

svm = SVC()
models.append(('SVM', svm))

ann = MLPClassifier()
models.append(('MLP', ann))
```

```python
from sklearn.ensemble import VotingClassifier

estimators = []

model0 = LogisticRegression()
estimators.append(('log', model0))

model1 = GaussianNB()
estimators.append(('nb', model1))

model2 = SVC()
estimators.append(('svm', model2))

model3 = MLPClassifier()
estimators.append(('ann', model3))

vot = VotingClassifier(estimators)
models.append(('VOT', vot))
```

```python
from sklearn.ensemble import BaggingClassifier

baglog = BaggingClassifier(base_estimator=LogisticRegression(), n_estimators=10)
models.append(('BAGLR', baglog))

bagnb = BaggingClassifier(base_estimator=GaussianNB(), n_estimators=10)
models.append(('BAGNB', bagnb))

bagsvm = BaggingClassifier(base_estimator=SVC(), n_estimators=10)
models.append(('BAGSVM', bagsvm))

bagann = BaggingClassifier(base_estimator=MLPClassifier(), n_estimators=10)
models.append(('BAGMLP', bagann))
```

```python
from sklearn.ensemble import AdaBoostClassifier

adalog = AdaBoostClassifier(base_estimator=LogisticRegression(), n_estimators=10)
models.append(('ADALR', adalog))

adanb = AdaBoostClassifier(base_estimator=GaussianNB(), n_estimators=10)
models.append(('ADANB', adanb))

adasvm = AdaBoostClassifier(base_estimator=SVC(probability=True), n_estimators=10)
models.append(('ADASVM', adasvm))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

results = []
names = []
msgs = []
scoring = 'accuracy'

for name, model in models:
  kfold = KFold(n_splits = 10)
  cv_results = cross_val_score(model, X, Y, cv = kfold, scoring = scoring)
  results.append(cv_results)
  names.append(name)
  msg = "%s:\t %f (%f)" % (name, cv_results.mean(), cv_results.std())
  msgs.append(msg)
```

```python
for accuracy in msgs:
    print(accuracy)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 6))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.4, random_state = 42)

adanb_model = AdaBoostClassifier(base_estimator=GaussianNB(), n_estimators=10)
adanb_model.fit(X_train, y_train)

result = adanb_model.score(X_test, y_test)
print(("Accuracy : %.3f%%") % (result * 100.0))
```
